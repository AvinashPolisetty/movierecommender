```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
movies=pd.read_csv('tmdb_5000_movies.csv')
credits=pd.read_csv('tmdb_5000_credits.csv')
```

```python
movies.head(2)
```

```python
credits.head(2)
```

```python
movies=movies.merge(credits,on='title')
```

```python
movies.head(3)
```

```python
movies.shape
```

```python
movies=movies[['movie_id','title','overview','genres','keywords','cast','crew']]
movies.head(2)
```

```python
movies.isnull().sum()
```

```python
movies.dropna(inplace=True)
```

```python
movies.duplicated().sum()
```

```python
movies['genres'][0]
```

```python
import ast
ast.literal_eval('[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]')
```

```python
##from the genres we have to take the 'name' from the genres


def convert(obj):
    l=[]
    for i in ast.literal_eval(obj):
        l.append(i['name'])
    return l
```

```python
movies['genres']=movies['genres'].apply(convert)
```

```python
movies.head(2)
```

```python
movies['keywords']=movies['keywords'].apply(convert)
```

```python
movies.head(2)
```

```python
ast.literal_eval(movies['cast'][0])
```

```python
#by using this function we are taking the only 3 cast members

def convert2(obj):
    l=[]
    counter=0
    for i in ast.literal_eval(obj):
        if counter < 3:
            l.append(i['name'])
            counter+=1
        else:
            break
    return l
```

```python
movies['cast']=movies['cast'].apply(convert2)
```

```python
movies.head(2)
```

```python
#By using this function we are taking only directors

def fetch_director(obj):
    l=[]
    for i in ast.literal_eval(obj):
        if (i['job']=='Director'):
            l.append(i['name'])
            break
    return l
```

```python
movies['crew']=movies['crew'].apply(fetch_director)
```

```python
movies.head(2)
```

```python
movies['overview']=movies['overview'].apply(lambda x: x.split())
```

```python
movies.head(2)
```

```python
movies['genres']=movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
movies['keywords']=movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
movies['cast']=movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
movies['crew']=movies['crew'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies.head(3)
```

```python
movies['tags']=movies['overview']+movies['genres']+movies['keywords']+movies['cast']+movies['crew']
```

```python
movies.sample(4)
```

```python
new_df=movies[['movie_id','title','tags']]
new_df.head()
```

```python
new_df['tags']=new_df['tags'].apply(lambda x : " ".join(x))
```

```python
new_df.head()
```

```python
new_df['tags']=new_df['tags'].apply(lambda x:x.lower())
```

```python
import nltk
```

```python
from nltk.stem import PorterStemmer
ps=PorterStemmer()
```

```python
def stemming(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
        
    return " ".join(y)
```

```python
new_df['tags']=new_df['tags'].apply(stemming)
```

```python
new_df.head()
```

```python
#coverting the words into the vectors

from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=4500,stop_words='english')
```

```python
vectors=cv.fit_transform(new_df['tags']).toarray()
vectors
```

```python
vectors.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity

similarity=cosine_similarity(vectors)
```

```python
similarity[2]
```

```python
def recommend(movie):
    index = new_df[new_df['title'] == movie].index[0]
    distance=similarity[index]
    movie_list = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1])[1:6]
    
    for i in movie_list:
        print(new_df.iloc[i[0]].title)
        
    
```

```python
recommend('Batman Begins')
```
